Exploración de correlación de datos y complejidad. Exclusión de outliers, separación de la variable suma_N. Primero con Standard, ahora 4 componentes.

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("Effor_index_MASTER.csv")
```

```python
mapa_fijo = {
    "nivel 1": 1,
    "nivel 2": 2,
    "nivel 3.1": 3,
    "nivel 3.2": 4,
    "nivel 3.3": 5,
    "nivel 3.4": 6,
    "nivel 3.5": 7,
    "nivel 3.6": 8,
    "nivel 3.7": 9,
    "nivel 3.8": 10,
    "nivel 3.9": 11,
    "nivel 4.1": 12,
    "nivel 4.2": 13,
    "nivel 4.3": 14
}

df["nivel_digitalizacion_numT"] = (
    df["nivel_digitalizacion"]
    .str.lower()
    .str.strip()
    .map(mapa_fijo)
)
```

```python
df["nivel_digitalizacion_num"] = df["nivel_digitalizacion_numT"].fillna(0)
```

```python
# Columna de existencia de costo
df["costo"] = df["TraCosto"].map({"VERDADERO": 1, "FALSO": 0})
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df[[
    "N_FORMATOS_FINAL",
    "CONTEO_NETO",
    "Tiempo_en_minutos",
    "nivel_digitalizacion_num",
    "costo"
]].dropna()

# Escalado
scaler_std = StandardScaler()
X_std = scaler_std.fit_transform(X)

# PCA
pca_std = PCA(n_components=3)
components_std = pca_std.fit_transform(X_std)

# Guardar resultados
var_std = pca_std.explained_variance_ratio_
comp_std = pca_std.components_

print("Varianza Standard:")
print(var_std)

print("\nComponentes Standard:")
print(comp_std)

# Guardar PCs
df.loc[X.index, "PC1_std"] = components_std[:, 0]
df.loc[X.index, "PC2_std"] = components_std[:, 1]
df.loc[X.index, "PC3_std"] = components_std[:, 2]
```

Ahora Robust:

```python
from sklearn.preprocessing import RobustScaler

scaler_rob = RobustScaler()
X_rob = scaler_rob.fit_transform(X)

pca_rob = PCA(n_components=3)
components_rob = pca_rob.fit_transform(X_rob)

# Guardar resultados
var_rob = pca_rob.explained_variance_ratio_
comp_rob = pca_rob.components_

print("\nVarianza Robust:")
print(var_rob)

print("\nComponentes Robust:")
print(comp_rob)

# Guardar PCs
df.loc[X.index, "PC1_rob"] = components_rob[:, 0]
df.loc[X.index, "PC2_rob"] = components_rob[:, 1]
df.loc[X.index, "PC3_rob"] = components_rob[:, 2]
```
